IMPORTING THE LIBRARIES

```python
import pandas as pd
import matplotlib as plt
import numpy as np
import seaborn as sns
```

IMPORTING THE DATASET

```python
bird = pd.read_csv(r'C:\Users\User\OneDrive\Desktop\Arewa DS&ML\ArewaDS-Machine-Learning\data\birds.csv')
```

```python
print(bird.head())
```

FILTER FOR SNOW GEESE

```python
snow_geese = bird[bird['Name'].str.contains('Snow Goose', case=False)].copy()
```

```python
print(snow_geese.head())
```

```python
snow_geese.shape
```

VISUALIZATIONS

```python
# Line chart: Show the range of Snow Goose body mass and length
plt.pyplot.figure(figsize=(10, 5))
plt.pyplot.plot(['MinLength', 'MaxLength'], [snow_geese['MinLength'].values[0], snow_geese['MaxLength'].values[0]], marker='o', label='Length (cm)')
plt.pyplot.plot(['MinBodyMass', 'MaxBodyMass'], [snow_geese['MinBodyMass'].values[0], snow_geese['MaxBodyMass'].values[0]], marker='o', label='Body Mass (g)')
plt.pyplot.title('Snow Goose: Length and Body Mass Range')
plt.pyplot.ylabel('Value')
plt.pyplot.legend()
plt.pyplot.show()

# Scatterplot: Compare Snow Goose with other geese in body mass vs wingspan
geese = bird[bird['Name'].str.contains('goose', case=False)]
plt.pyplot.figure(figsize=(8, 6))
sns.scatterplot(data=geese, x='MaxWingspan', y='MaxBodyMass', hue='Name', s=100)
plt.pyplot.scatter(snow_geese['MaxWingspan'], snow_geese['MaxBodyMass'], color='red', label='Snow Goose', edgecolor='black', s=200)
plt.pyplot.title('Geese: Max Wingspan vs Max Body Mass')
plt.pyplot.xlabel('Max Wingspan (cm)')
plt.pyplot.ylabel('Max Body Mass (g)')
plt.pyplot.legend()
plt.pyplot.show()

# Bar chart: Show all measured attributes for Snow Goose
attributes = ['MinLength', 'MaxLength', 'MinBodyMass', 'MaxBodyMass', 'MinWingspan', 'MaxWingspan']
values = [snow_geese[attr].values[0] for attr in attributes]
plt.pyplot.figure(figsize=(10, 5))
sns.barplot(x=attributes, y=values, palette='Blues')
plt.pyplot.title('Snow Goose: Physical Attributes')
plt.pyplot.ylabel('Value')
plt.pyplot.show()
```
